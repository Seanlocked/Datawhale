```python
# 挑出object类型，便于后面编码 
cat_vars = []
def fea_categorical_check(df):
    print('描述变量有:\n')
    for col in df.columns:
        if df[col].dtype == 'object':
            print(col)
            cat_vars.append(col)
    return cat_vars
    
```

```python
fea_categorical_check(data)
```

```python
# 移除特征，检查每个特征的取值数 ，删除单值特征 
def fea_value_cnt_check(df):
    single_fea = []
    for col in df.columns:
        cnt = df[col].value_counts().count() # value_counts()之后再count()，数出有多少项目
        print('{col}的取值有{cnt}个'.format(col=col,cnt=cnt))
        if cnt == 1:
            single_fea.append(col)
    return single_fea
            
        
```

```python
fea_value_cnt_check(data)
```

```python
# isnull()是显示整个df中的缺失值，出现为逻辑值。加sum()计算出每一列中的缺失值个数 
fea_null_sum = data.isnull().sum() 
```

```python
# 通过np.argmax()方法求出缺失最多的特征，选择丢弃 
np.argmax(fea_null_sum)
```

```python
# 观察obejct类型的值分布 
for col in cat_vars:
    print('描述型变量{col}的分布为:'.format(col=col))
    print(data[col].value_counts())
    print('\n')
    
```

```python
# 丢弃缺失特征最多的student_feature,单值特征'bank_card_no', 'source' ，描述型特征trade_no,id_name,latest_query_time,loans_latest_time
del_col = ['student_feature','bank_card_no', 'source','trade_no','id_name','latest_query_time','loans_latest_time','Unnamed: 0','custid']
df_1 = data.drop(columns=del_col)
```

```python
# 将reg_preference_for_trad 进行编码转化为数值，致谢19-edward代码
map_dict={'一线城市':1,'二线城市':2,'三线城市':3,'境外':4,'其他':0}
df_1['reg_preference_for_trad']=df_1['reg_preference_for_trad'].map(map_dict)
df_1.head()
```

```python
# 取出目标值，再对剩下的特征进行均值填补以及标准化 
y = df_1['status']
df_1 = df_1.drop(columns=['status'],axis=1)
```

```python
# 缺失值采用均值填补，较为粗糙的解决办法...
from sklearn.preprocessing import Imputer
imputer = Imputer(strategy='median')
df_2 = imputer.fit_transform(df_1) # 出来的结果是array类型 
df_2 = pd.DataFrame(df_2,columns=df_1.columns) # 转换为df格式，将列赋值上去 
```

```python
df_2.head()
```

```python
# 检查初步清洗后的数据是否还有缺失值 df_2.isnull().sum()[bool] 取出有缺失的数据，结果为空，说明无空值了
df_2.isnull().sum()[df_2.isnull().sum() != 0 ]
```

```python
# 将数据进行标准化，作为模型训练前的最后处理 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_3 = scaler.fit_transform(df_2) # 生成结果是array类型，需要转化为df类型
df_3 = pd.DataFrame(df_3,columns=df_2.columns)
```

```python
df_3.head()
```

```python
# 数据清洗且初步处理完成，下面切分数据进行训练预测
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test = train_test_split(df_3,y,test_size=0.3,random_state=2018)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# 使用svm训练模型
from sklearn.svm import SVC
clf = SVC()
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
clf.score(X_test,y_test)
```

当前svm模型评分为0.780

```python
from sklearn.metrics import recall_score
recall_score(y_test,y_pred)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test,y_pred)
```
